```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.use('TkAgg')


plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# ── 데이터 로드 ──────────────────────────────
train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")

print(f"train.shape: {train.shape} test.shape: {test.shape}")
train.head()
```

```python
# ── 기본 정보 ─────────────────────────────────
train.info()
train.describe()
# 결측치 확인
print("=== 결측치 현황 ===")
missing = train.isnull().sum()
print(missing[missing > 0])
```

```python
# ── 생존율 전체 ───────────────────────────────
print(f"전체 생존율 : {train['Survived'].mean().round(2)}")
```

```python
# ── 성별 생존율 ───────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

train.groupby('Sex')['Survived'].mean().plot(
    kind='bar', ax=axes[0], title='성별 생존율', color=['#E24B4A','#378ADD']
)
axes[0].set_xlabel('')
axes[0].tick_params(axis='x', rotation=0)
axes[0].set_ylabel('생존율')

# ── 객실등급 생존율 ────────────────────────────
train.groupby('Pclass')['Survived'].mean().plot(
    kind='bar', ax=axes[1], title='객실등급 생존율', color=['#1D9E75','#5DCAA7','#9FE1CB']
)
axes[1].set_xlabel('객실등급')
axes[1].tick_params(axis='x', rotation=0)

# ── 나이 분포 ─────────────────────────────────
train[train['Survived']==1]['Age'].dropna().plot(
    kind='hist', ax=axes[2], alpha=0.6, label='생존', bins=25, color='#378ADD'
)
train[train['Survived']==0]['Age'].dropna().plot(
    kind='hist', ax=axes[2], alpha=0.6, label='사망', bins=25, color='#E24B4A'
)
axes[2].set_title('나이별 생존/사망 분포')
axes[2].legend()

plt.tight_layout()
plt.savefig('../outputs/eda_overview.png', dpi=150)
plt.show()
```

```python
# ── 상관관계 히트맵 ────────────────────────────
plt.figure(figsize=(8, 6))
numeric_cols = ['Survived','Pclass','Age','SibSp','Parch','Fare']
sns.heatmap(train[numeric_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('변수 간 상관관계')
plt.tight_layout()
plt.show()

# ── 호칭(Title) 분석 ─────────────────────────
train['Title'] = train['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)
print(train.groupby('Title')['Survived'].agg(['mean','count']))
```
